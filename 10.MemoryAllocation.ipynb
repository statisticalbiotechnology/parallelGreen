```python
from utils2 import *
import time
from significance_of_mean_cuda import significance_of_mean_cuda
```

### Get synthetic data

```python
path = "figures/memoryAllocation"
```

```python
num_examples  = 1_000
```

```python
N = list(range(10,150,10))
s = [10,20,30]
```

```python
allMemNList = list()
for bins in s:
    memList = list()
    for n in N:
        A = np.asarray([np.random.normal(0, 1, n) for _ in range(num_examples)])
        B = np.asarray([np.random.normal(1, 1, n) for _ in range(num_examples)])
        
        SGM = significance_of_mean_cuda(bins,dtype_v=np.uint16,dtype_A=np.float64)
        SGM.run(A,B)
        
        memList.append(SGM.mem_alloc)
        
    allMemNList.append(memList)
        
```

```python
plotSNS((allMemNList[0],r"$n_{w}=10$"), (allMemNList[1],r"$n_{w}=20$"), (allMemNList[2],
                                        r"$n_{w}=30$") , N, "Memory Allocation(MiB)", r"$n$", log=False,path=getPATH(path, "N"))
```
